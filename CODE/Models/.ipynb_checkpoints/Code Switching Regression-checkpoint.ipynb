```python
#mod = smf.ols(formula='Lottery ~ Literacy + Wealth + Region', data=df)
#mod = sm.Logit(formula='"has naija" ~ depth', data=df)
y, X = patsy.dmatrices('has_naija ~ depth + C(day_type) + C(month) + sentiment '
                       '+ parent_sentiment + C(origin) + parent_mentions_tribe', df, return_type='dataframe')
mod = sm.Logit(y, X)
res = mod.fit()
print(res.summary())
```
